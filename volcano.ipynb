```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# EVENT DETAILS

df = pd.read_csv('/content/drive/MyDrive/TSWD Final Project/volcano/events.csv')
df.head()
```

```python
# LIST OF EVENTS AND THEIR FREQUENCIES

eventTypes = sorted(list(set(df.event_type.unique())))

count = {}
for i in df.event_type:
  if i in count:
    count[i] += 1
  else:
    count[i] = 1

sortedDict = sorted(count.items(), key = lambda x: x[1])
for i in sortedDict:
  print(i[0], ":", i[1])
```

```python
# TOTAL NUMBER OF EVENTS

print(len(sortedDict))
```
